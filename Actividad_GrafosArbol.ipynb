**Árbol Binario**

```python
# Crear la clase Nodo, será como una rama del arbol binario
class Nodo:
    def __init__(self, dato):
        # Aquí guardamos el dato (valor) que lleva el nodo 
        self.dato = dato
        # Izq. es el hijo izquierdo, son siempre números menores que el papá o raiz.
        self.izq = None
        # der. es el hijo derecho, son siempre números mayores que el papá/raiz
        self.der = None
```

```python
# Crear la clase ArbolBinarioo
class ArbolBinario:
    def __init__(self):
        # La raiz es el nodo principal y es None porque al inicio no hay valores.
        self.raiz = None

    # Método para insertar un nuevo vallor
    def insertar(self, dato):
        # Si el arbol está vacío, el nuevo dato insertado es la raiz del árbol
        if self.raiz is None:
            self.raiz = Nodo(dato)
            print(f"Dato inertado {dato} como raiz.") # Mensaje para confirmar que se ha insertado el nuevo dato
        else:
            # Si ya hay un dato en la raiz, usamos la función insertar_recursivo
            self._insertar_recursivo(dato, self.raiz)

    # Método recursivo, es decir se llama a sí misma
    def _insertar_recursivo(self, dato, nodo_actual):
        #  Si el dato es menor que el nodo actual, se va a la izquierda
        if dato < nodo_actual.dato:
            # Si no hay nada a a izquierda entonces ese dato se inserta alli a la izquierda
            if nodo_actual.izq is None:
                nodo_actual.izq = Nodo(dato)
                print(f" {dato} -> Insertado a la izquierda.")
            else:
                # Si ya había dato allí en la izquierda, entonces se sigue bajando por la izquierda.
                self._insertar_recursivo(dato, nodo_actual.izq)

        # Si el dato es mayor que el nodo actual, se va a la derecha 
        elif dato > nodo_actual.dato:
            # Si el nodo actual de la derecha está vacío,  entonces se inserta allí a la derecha.
            if nodo_actual.der is None:
                nodo_actual.der = Nodo(dato)
                print(f" {dato} -> Insertado a la derecha.")
            else:
                # Si no estaba vacío, sigue bajando por la derecha
                self._insertar_recursivo(dato, nodo_actual.der)

        else:
            print("El dato ya existe, no lo ingresamos 2 veces")

    # Método para buscar un valor
    def buscar(self, dato):
        # LLamamos a la función recursiva desde la raiz
        return self._buscar_recursivo(dato, self.raiz)
    
    # Función recursiva para buscar 
    def _buscar_recursivo(self, dato, nodo):
        # Si el nodo is None, retornamos False porque no existe
        if nodo is None:
            return False
        # Si el dato buscado es igual al nodo (raiz), retorna True 
        if dato == nodo.dato:
            return True
        # Si el dato buscado es menor que el nodo(raiz), se busca en el subarbol de la izquierda
        if dato < nodo.dato:
            return self._buscar_recursivo(dato, nodo.izq)
        # Si no está por la izquierda se busca por el subarbol de la derecha
        return self._buscar_recursivo(dato, nodo.der)
    
    # RECORRIDOS
    # inOrden: izquierda -> raiz -> derecha (aquí debe salir ordenado)
    def inOrden(self):
        resultado = [] # Lista para guardar los valores en orden
        self._inOrden_recursivo(self.raiz, resultado)
        return resultado
    
    # Función recursiva del inOrden
    def _inOrden_recursivo(self, nodo, resultado):
        # Si hay un nodo(raiz), entonces se va por todo lo de la izquierda
        if nodo:
            # Izquierda, son los menores
            self._inOrden_recursivo(nodo.izq, resultado)
            # El nodo actual (raiz) 
            resultado.append(nodo.dato)
            # Derecha, son los mayores
            self._inOrden_recursivo(nodo.der, resultado)
        # Debe imprimirse en orden ascendente
```

```python
# Se crea a clase Grafo
class Grafo:
    def __init__(self):
        # El diccionario, donde la clave es el vertice y el valor va a ser la lista de vecinos o aristas.
        self.vertices = {} 

    # Método agregar veertice sin aristas
    def agregar_vertice(self, vertice):
        # Si no existe lo creamos con lista vacía de vecinos
        if vertice not in self.vertices:
            self.vertices[vertice] = []
            print(f"Vertice {vertice} agregado.")
        else:
            print("El {vertice} ya está")

    # Agregar arista: dirigida de origen a destino
    def agregar_arista(self, origen, destino):
        # Hay que asegurarse de que ambos vertices existan
        if origen not in self.vertices:
            self.agregar_vertice(origen)
        if destino not in self.vertices:
            self.agregar_vertice(destino)
        # Si no hay arista ya entoncesd la agregamos
        if destino not in self.vertices[origen]:
            self.vertices[origen].append(destino)
            print(f"Arista {origen} --> {destino} creada")
        else:
            print(f"Ya existía arista {origen} --> {destino} ")

    # búsqueda en profundidad (DFS)
    def DFS(self, vertice_inicio):
        # Marcamos los visitados, para no repetirlos
        visitados = set() # Crea un set vacío ara guardar los que ya se visitaron
        resultado = []
        # Llamamos a la función recursiva
        self._DFS_recursivo(vertice_inicio, visitados, resultado)
        return resultado
    
    # Función recursiva del DFS
    def _DFS_recursivo(self, vertice_actual, visitados, resultado):
        # Si ya vimos ese vertice entonces se salta a otro
        if vertice_actual in visitados:
            return
        # Agregarlo como visitado 
        visitados.add(vertice_actual)
        resultado.append(vertice_actual)
        print(f"Visitado: {vertice_actual}") # Imprime mientras se ejecuta

        # Recorremos todos los vecinos o aristas
        for vecino in self.vertices.get(vertice_actual, []):
            # Se llama a la función recursiva
            self._DFS_recursivo(vecino, visitados, resultado)    
```
